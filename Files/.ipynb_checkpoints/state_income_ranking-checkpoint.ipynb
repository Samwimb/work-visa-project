```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "state_income_ranking_2017.csv"
income_2017_df = pd.read_csv(csv_file)
income_2017_df.head()
```

```python
csv_file = "state_income_ranking_2016.csv"
income_2016_df = pd.read_csv(csv_file)
income_2016_df.head()
```

```python
cleaned_income_2017_df = income_2017_df[['GRT_STUB.rank-label', 'GRT_STUB.display-label', 'EST', 'MOE']].copy()
cleaned_income_2017_df.head()
```

```python
cleaned_income_2016_df = income_2016_df[['GRT_STUB.rank-label', 'GRT_STUB.display-label', 'EST', 'MOE']].copy()
cleaned_income_2016_df.head()
```

```python
cleaned_income_2017_df.rename(columns={'GRT_STUB.rank-label':'Rank',
                          'GRT_STUB.display-label':'State',
                          'EST': 'Median Household Income',
                            'MOE': 'Margin of Error'}, 
                 inplace=True)
cleaned_income_2017_df.head()
```

```python
cleaned_income_2017_df.drop([0], inplace=True)
```

```python
cleaned_income_2017_df.head()
```

```python
cleaned_income_2017_df.dropna(inplace=True)
```

```python
cleaned_income_2017_df
```

```python
cleaned_income_2016_df.rename(columns={'GRT_STUB.rank-label':'Rank',
                          'GRT_STUB.display-label':'State',
                          'EST': 'Median Household Income',
                            'MOE': 'Margin of Error'}, 
                 inplace=True)
cleaned_income_2016_df.head()
```

```python
cleaned_income_2016_df.drop([0], inplace=True)
```

```python
cleaned_income_2016_df.head()
```

```python
cleaned_income_2016_df.dropna(inplace=True)
```

```python
cleaned_income_2016_df
```

```python
cleaned_income_2016_df['Year'] = 2016
```

```python
cleaned_income_2017_df['Year'] = 2017
```

```python
income = [cleaned_income_2016_df, cleaned_income_2017_df]
merged_income = pd.concat(income, sort=False)
```
