```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "state_income_ranking_2017.csv"
income_2017_df = pd.read_csv(csv_file)
income_2017_df.head()
```

```python
csv_file = "state_income_ranking_2016.csv"
income_2016_df = pd.read_csv(csv_file)
income_2016_df.head()
```

```python
cleaned_income_2017_df = income_2017_df[['GRT_STUB.rank-label', 'GRT_STUB.display-label', 'EST', 'MOE']].copy()
cleaned_income_2017_df.head()
```

```python
cleaned_income_2016_df = income_2016_df[['GRT_STUB.rank-label', 'GRT_STUB.display-label', 'EST', 'MOE']].copy()
cleaned_income_2016_df.head()
```
